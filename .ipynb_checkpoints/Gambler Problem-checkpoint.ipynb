```python
import numpy as np
```

```python
S = np.arange(1,100)  # different states
```

```python
p = 0.4 # probability of heads
```

```python
def get_available_actions(s):
    top = min(s, 100-s)
    return np.arange(1, top+1)
```

```python
pi = [None]
for i in range(1, 100):
    A = get_available_actions(i)
    pi.append(np.ones(A)/len(A))
```

```python
pi
```

```python
def value_iteration(gamma, theta):
    V = np.zeros(len(S)+1)
    while True:
        delta = 0
        for s in S:
            v = V[s]
            A = pi[s]
            bellman_optimality_update(A, V, s, gamma)
            delta = max(delta, abs(v - V[s]))
        if delta < theta:
            break
    for s in S:
        A = get_available_actions(s)
        q_greedify_policy(A, V, pi, s, gamma)
    return V, pi
```

```python
def bellman_optimality_update(A, V, s, gamma):
    a = np.zeros(len(A))
    for i, action in enumerate(A):
        t = 0
        trans = transitions(s, action)
        for s_, (r, p) in enumerate(trans):
            t += p*(r + gamma*V[s_])        
        a[i] = t
    V[s] = np.max(a)
```

```python
def q_greedify_policy(A, V, pi, s, gamma):
    a = np.zeros(len(A))
    for i, action in enumerate(A):
        t = 0
        trans = transitions(s, action)
        for s_, (r, p) in enumerate(trans):
            t += p*(r + gamma*V[s_])
        a[i] = t
    j = np.argmax(a)
    a = np.zeros(len(A))
    a[j] = 1
    pi[s] = a
```

```python
def transitions(s, action):
    s_1 = s + action
    s_2 = s - action
    r_1 = 1 if s_1 == 100 else 0
    r_2 = 0
    p_1 = p
    p_2 = 1 - p
    return [(r_1, p_1), (r_2, p_2)]
```

```python
V_opt, pi_opt = value_iteration(1, 0.1)
```
